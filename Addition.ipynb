```python
import numpy as np

from os import listdir
from scipy.io import wavfile


from datetime import datetime
from numba import njit

from matplotlib import pyplot as plt
```

```python
def ls(path):
    array = listdir(path)
    ans = []
    for i in array:
        if i.startswith('.') == False:
            ans.append(i)
    return ans
```

```python
ls('musics')
```

```python
music = wavfile.read('musics/' + ls('musics')[0])
x = np.array(music[1],dtype=float)
x = np.squeeze(x)[:10000]
```

```python
time = np.linspace(0., x.shape[0] / music[:len(x)][0], x.shape[0])
plt.figure(figsize=(15, 4))
plt.plot(time, x[:,1])
#plt.xticks(np.arange(0, len(x)+1, 1000))
plt.show()
```
